```python
import numpy as np
```

```python
matriz_confusao = np.array([[50, 10],  # Verdadeiros positivos (VP) e Falsos positivos (FP)
                            [5,  35]]) # Falsos negativos (FN) e Verdadeiros negativos (VN)

print("Matriz de Confusão:")
print(matriz_confusao)
```

```python
# Funções para cálculo das métricas
def sensibilidade(VP, FN):
    return VP / (VP + FN)

def especificidade(VN, FP):
    return VN / (VN + FP)

def acuracia(VP, VN, N):
    return (VP + VN) / N

def precisao(VP, FP):
    return VP / (VP + FP)

def f_score(precisao, sensibilidade):
    return 2 * (precisao * sensibilidade) / (precisao + sensibilidade)
```

```python
# Extração dos valores da matriz de confusão
VP = matriz_confusao[0, 0]
FP = matriz_confusao[0, 1]
FN = matriz_confusao[1, 0]
VN = matriz_confusao[1, 1]

# Número total de elementos
N = matriz_confusao.sum()

# Cálculo das métricas
sens = sensibilidade(VP, FN)
espec = especificidade(VN, FP)
acc = acuracia(VP, VN, N)
prec = precisao(VP, FP)
f1 = f_score(prec, sens)

# Exibir os resultados
print("\nMétricas calculadas:")
print(f"Sensibilidade (Recall): {sens:.2f}")
print(f"Especificidade: {espec:.2f}")
print(f"Acurácia: {acc:.2f}")
print(f"Precisão: {prec:.2f}")
print(f"F1-Score: {f1:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotar a matriz de confusão
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusao, annot=True, cmap="Blues", fmt="d",
            xticklabels=["Classe Positiva", "Classe Negativa"],
            yticklabels=["Classe Positiva", "Classe Negativa"])
plt.xlabel("Previsto")
plt.ylabel("Verdadeiro")
plt.title("Matriz de Confusão")
plt.show()
```
